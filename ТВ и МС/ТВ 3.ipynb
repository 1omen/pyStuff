```python
def ExpValue (x, p): #функция для вычисления математического ожидания, на вход получает таблицу распределения
    return np.sum(x*p)
```

**Шаг 2**

```python
def ExpValueFun (f, xi, pi): #математическое ожидание при применении функции к таблице распределения
    xi = np.array(list(map(lambda x:f(x), xi)))
    return ExpValue(xi,pi)

def FunExp (f, x, p): #выполнение заданной функции от математического ожидания
    return f(ExpValue(x,p))
```

# Задание 2

**Шаг 1**

```python
# Все нижеописанные функции возвращают значения дисперсий и среднеквадратическое отклонение для определенных аргументов 
# в виде (Дисперсия, Отклонение)

# способ расчета дисперсии по формуле D(x)=M(x^2)−(M(x))^2

def Var_simple_1(x, p): # функция для расчета дисперсии для заданной таблицы
    var = ExpValue(x**2, p) - ExpValue(x,p)**2
    return var, np.sqrt(var) 
 
def Variance_1(f, xi, pi): # функция для расчета дисперсии при применении функции к таблице
    xi = np.array(list(map(lambda x:f(x), xi)))
    var = ExpValue(xi**2, pi) - ExpValue(xi,pi)**2
    return var, np.sqrt(var) 
```

**Шаг 2**

```python
# способ расчета дисперсии по формуле D(x)=M(x−M(x))^2

def Var_simple_2(x, p): # функция для расчета дисперсии для заданной таблицы
    var = ExpValue((x-ExpValue(x,p))**2,p)
    return var, np.sqrt(var)

def Variance_2(f, xi, pi): # функция для расчета дисперсии при применении функции к таблице
    xi = np.array(list(map(lambda x:f(x), xi)))
    var = ExpValue((xi-ExpValue(xi,pi))**2,pi)
    return var, np.sqrt(var)
```

**Шаг 3**

```python
# расчеты дисперсий для распределений заданных таблицами

print("Для таблицы 5, 1 формула", Var_simple_1(x5, p5))
print("Для таблицы 5, 2 формула", Var_simple_1(x5, p5))
print()
print("Для таблицы 79, 1 формула", Var_simple_1(x79, p79))
print("Для таблицы 79, 2 формула", Var_simple_1(x79, p79))
```

```python
# расчеты дисперсии для 1 функции

print("Для функции 5x^2+2, 5 таблица, 1 формула", Variance_1(f1, x5, p5))
print("Для функции 5x^2+2, 5 таблица, 2 формула", Variance_2(f1, x5, p5))
print()
print("Для функции 5x^2+2, 79 таблица, 1 формула", Variance_1(f1, x79, p79))
print("Для функции 5x^2+2, 79 таблица, 2 формула", Variance_2(f1, x79, p79))
```
